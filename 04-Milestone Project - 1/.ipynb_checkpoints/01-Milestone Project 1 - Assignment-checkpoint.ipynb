```python
import numpy as np
global row1,row2,row3,player_num

def player_symbol():
    player1_symbol = input('Hello Player 1. Please enter the symbol you want to take (X or O):  ')
    
    if player1_symbol not in ('X','O'):
        print('Sorry! Wrong choice')
        player1_symbol = input('Hello Player 1. Please enter the symbol you want to take (X or O):  ')
        
    return player1_symbol


def game_board():
    print('\n\n')
    print(row1)
    print(row2)
    print(row3)
    
def validate_input(row_num,col_num):
    print('Row: ',row_num,' Column: ',col_num)
    if int(row_num) in np.arange(1,4) and int(col_num) in np.arange(1,4):
        pass
    else:
        print('The row and column number entered is invalid')
        row_num = input(f'Player {player_num} enter row to place your symbol: ')
        col_num = input(f'Player {player_num} enter column to place your symbol: ')
        row_num,col_num = validate_input(row_num,col_num)
    
    return row_num,col_num
    
def assign_position(row,col,player1_symbol,player2_symbol):
    row_names = {1:'row1',2:'row2',3:'row3'}
    
    row_num = int(row)
    col_num = int(col)-1
    
    if globals()[row_names[row_num]][col_num] != '     ':
        print('The chosen position is already filled')
        row = input(f'Player {player_num} enter row to place your symbol: ')
        col = input(f'Player {player_num} enter column to place your symbol: ')
    else:
        pass
    
    if player_num == 1:
        globals()[row_names[row_num]][col_num] = '  '+player1_symbol+'  '
    else:
        globals()[row_names[row_num]][col_num] = '  '+player2_symbol+'  '
        
            
def check_win():
    game_won = False
    game_draw = False
    if ((row1[0] == row1[1] == row1[2] != '     ') or (row2[0] == row2[1] == row2[2] != '     ') or 
        (row3[0] == row3[1] == row3[2] != '     ') or (row1[0] == row2[0] == row3[0] != '     ') or 
        (row1[1] == row2[1] == row3[1] != '     ') or (row1[2] == row2[2] == row3[2] != '     ') or
        (row1[0] == row1[1] == row2[2] != '     ') or (row1[2] == row2[1] == row3[0] != '     ')):
        game_board()
        print(f'Congratulations player {player_num}. You won the game!')
        game_won = True
    else:
        game_won = False
    
    if (row1[0] != '     ' and row1[1] != '     ' and row1[2] != '     ' and
        row2[0] != '     ' and row2[1] != '     ' and row2[2] != '     ' and
        row3[0] != '     ' and row3[1] != '     ' and row3[2] != '     ' and
        game_won == False):
        game_board()
        print(f'The game is a draw. Try again next time!')
        game_draw = True
    else:
        game_draw = False
    
    return game_won, game_draw
    
```

```python
player1_symbol = player_symbol()

if player1_symbol == 'X':
    player2_symbol = 'O'
else:
    player2_symbol = 'X'
    
print(f'Player1 uses {player1_symbol} and player 2 uses {player2_symbol}')
print('Let us start the game!\n\n')

game_won = False
game_draw = False
row1 = ['     ','     ','     ']
row2 = ['     ','     ','     ']
row3 = ['     ','     ','     ']
player_num = 1


while game_won == False and game_draw == False:
    game_board()
    
    row_num = input(f'Player {player_num} enter row to place your symbol: ')
    col_num = input(f'Player {player_num} enter column to place your symbol: ')
    
    row_num,col_num = validate_input(row_num,col_num)
    
    assign_position(row_num,col_num,player1_symbol,player2_symbol)
    
    game_won, game_draw = check_win()
    
    if game_won == False:
        print(player_num)
        if player_num == 1:
            player_num = 2
        else:
            player_num = 1
    else:
        pass
```
